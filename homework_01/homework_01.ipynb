## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
number = input("введите число: ")
result = int(number[::-1])
print(result, type(result))
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
days_to_vacation = input("Введите количество дней до отпуска")
if days_to_vacation.isdigit() and int(days_to_vacation) >= 0:
    days_to_vacation = int(days_to_vacation)
    print(days_to_vacation // 7 * 2 + days_to_vacation % 7 // 5)
else:
    print("Вы ввели некорректное значение", days_to_vacation)
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
weight = int(input("Введите ширину плитки"))
long = int(input("Введите ширину плитки"))
piece = int(input("Введите размер куска, который хотите отломить"))
print(piece % weight == 0 and piece / weight <= long or piece % long == 0 and piece / long <= weight)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
number = int(input("Введите целое положительное число: "))
rome_number = ["M", "D", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
arabic_number = [1000, 500, 100, 90, 50, 40, 10, 9, 5, 4, 1]
index = 0
result = ""
while number != 0:
    result += rome_number[index] * (number // arabic_number[index])
    number = number % arabic_number[index]
    index += 1
print(result)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
float_number = input("Введите вещественное число:")
if_has_dot = False
is_float_number = True
is_negative = False
numbers = "0123456789"
if float_number[0] == '-':
    is_negative = True
    float_number = float_number[1:]
elif float_number[0] == '+':
    float_number = float_number[1:]
for i, char in enumerate(float_number):
    if char == '.':
        if if_has_dot:
            is_float_number = False
            break
        if_has_dot = True
    elif char not in numbers:
        is_float_number = False
        break

print("Данное число")
if is_float_number and is_negative:
    print("отрицательное вещественное")
elif is_float_number and not is_negative:
    print("положительное вещественное")
else:
    print("невещественное")
```
